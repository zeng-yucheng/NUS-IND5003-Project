## 1. Data cleaning

### 1.1 Training features

```python
# clean the crashed listings in csv file
# creat Listings_clean.csv
```

```python
import pandas as pd
df = pd.read_excel("Airbnb Data/Listings_clean.xlsx")
```

```python
df.head()
```

```python
X = pd.DataFrame()
X['host_since'] = (pd.Timestamp(2021, 6, 30) - df['host_since']).dt.days
X.head()
```

```python
print(df.host_response_time.value_counts())
host_response_time_dict = {'within an hour': 1, 'within a few hours': 12, 'within a day': 24, 'a few days or more': 72}
df['host_response_time'].fillna('within an hour', inplace=True)
X['host_response_time'] = df['host_response_time'].apply(lambda x: host_response_time_dict[x])
X.head()
```

```python
X['host_response_rate'] = df['host_response_rate']
X['host_response_rate'].fillna(X['host_response_rate'].median(), inplace=True)
X['host_acceptance_rate'] = df['host_acceptance_rate']
X['host_acceptance_rate'].fillna(X['host_acceptance_rate'].median(), inplace=True)
X.head()
```

```python
print(df.host_is_superhost.value_counts())
print(df.host_has_profile_pic.value_counts())
print(df.host_identity_verified.value_counts())
print(df.instant_bookable.value_counts())

X['host_is_superhost'] = df['host_is_superhost'].apply(lambda x: 1 if x == 't' else 0)
X['host_has_profile_pic'] = df['host_has_profile_pic'].apply(lambda x: 1 if x == 't' else 0)
X['host_identity_verified'] = df['host_identity_verified'].apply(lambda x: 1 if x == 't' else 0)
X['instant_bookable'] = df['instant_bookable'].apply(lambda x: 1 if x == 't' else 0)

X.head()
```

```python
X['host_total_listings_count'] = df['host_total_listings_count']
X['latitude'] = df['latitude']
X['longitude'] = df['longitude']
X['room_type'] = df['room_type']
X['accommodates'] = df['accommodates']
X.head()
```

```python
X['bedrooms'] = df['bedrooms']
X['bedrooms'].fillna(1, inplace=True)
X['bedrooms'] = X['bedrooms'].astype(int)
X.head()
```

```python
X['price'] = df['price']
X['minimum_nights'] = df['minimum_nights']
X['maximum_nights'] = df['maximum_nights']
X.head()
```

```python
X['amenities'] = df['amenities'].apply(lambda x: x[1: -1].replace(' ', '').replace('"', '').split(','))
X.head()
```
